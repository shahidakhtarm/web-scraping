## 2. Data Cleaning

```python
import pandas as pd
import numpy as np
import os
dataset = pd.read_csv("ds2.csv", encoding='latin')   
dataset.head(5)
```

```python
dataset['yearly_change'] = dataset['yearly_change'].str.strip('%')
dataset['world_share'] = dataset['world_share'].str.strip('%')
dataset['urbon_pop'] = dataset['urbon_pop'].str.strip('%')
dataset.head(5)
```

```python
dataset['population'] = dataset['population'].str.replace(',', '')
dataset['net_change'] = dataset['net_change'].str.replace(',', '')
dataset['land_area'] = dataset['land_area'].str.replace(',', '')
dataset['migrants'] = dataset['migrants'].str.replace(',', '')
dataset['density'] = dataset['density'].str.replace(',', '')
dataset.head()
```

```python
def div(data):
    return float(data)/100
```

```python
def urban(data):
    try:
        return float(data)/100
    except:
        return None
```

```python
dataset['yearly_change'] = dataset['yearly_change'].apply(div)
dataset['world_share'] = dataset['world_share'].apply(div)
dataset['urbon_pop']=dataset['urbon_pop'].apply(urban)
```

```python
def ftrt(data):
     try:
        return float(data)
     except:
        return None
dataset['fert_rate']=dataset['fert_rate'].apply(ftrt)
```

```python
def age(data):
     try:
        return int(data)
     except:
        return None
dataset['med_age']=dataset['med_age'].apply(age)
```

```python
def mig(data):
     try:
        return int(data)
     except:
        return None
dataset['migrants']=dataset['migrants'].apply(mig)
```

```python
def den(data):
     try:
        return int(data)
     except:
        return None
dataset['density']=dataset['density'].apply(den)
```

```python
def den1(data):
     try:
        return float(data)
     except:
        return None
dataset['world_share']=dataset['world_share'].apply(den1)
```
