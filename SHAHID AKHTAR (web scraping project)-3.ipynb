# 3. Uni varient Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
import pandas as pd
import numpy as np
import os
import cv2
```

```python
data = pd.read_csv("ds3.csv",encoding='latin')
data.head()
```

```python
data1 = data.copy()
```

### population

```python
data1.set_index('country_name',inplace=True)
```

```python
data1.head() # top 5 countries have High population
```

```python
data1.tail() # top 5 countries have low population
```

```python
fig=plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
data1['population'].head(5).plot('pie')
plt.title('top 5')
plt.subplot(2,2,2)
data1['population'].tail(5).plot('pie')
plt.title('least 5')
fig.savefig('shaid1.png')
```

```python
fig=plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
data1['population'].head(5).plot('bar')
plt.title('top 5')
plt.subplot(2,2,2)
data1['population'].tail(5).plot('bar')
plt.title('least 5')
fig.savefig('shaid2.png')
```

```python
fig=plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
data1['population'].head(5).plot('box')
plt.title('top 5')
plt.subplot(2,2,2)
data1['population'].tail(5).plot('box')
plt.title('least 5')
fig.savefig('shaid3.png')
```
